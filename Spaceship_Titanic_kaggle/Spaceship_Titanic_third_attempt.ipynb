## Processing data

```python
train_df = raw_train.set_index('PassengerId')
test_df = raw_test.set_index('PassengerId')
```

```python
train_df.columns = list(map(lambda x: x.lower(), list(train_df.columns)))
test_df.columns = list(map(lambda x: x.lower(), list(test_df.columns)))
```

```python
train_df.drop('name', axis = 1, inplace = True)
test_df.drop('name', axis = 1, inplace = True)
```

```python
train_df.head()
```

### Changing object values to numeric

```python
train_df.homeplanet.value_counts()
```

```python
homeplanet_dict = {
    'Earth': 0,
    'Europa': 1,
    'Mars': 2
}

train_df['homeplanet'] = train_df.homeplanet.map(homeplanet_dict)
test_df['homeplanet'] = test_df.homeplanet.map(homeplanet_dict)
```

```python
train_df.info()
```

```python
train_df.destination.value_counts()
```

```python
destination_dict = {
    'TRAPPIST-1e': 0,
    '55 Cancri e': 1,
    'PSO J318.5-22': 2
}

train_df['destination'] = train_df.destination.map(destination_dict)
test_df['destination'] = test_df.destination.map(destination_dict)
```

```python
train_df.info()
```

```python
train_df['vip'] = train_df.vip.astype('float64')
test_df['vip'] = test_df.vip.astype('float64')

train_df['cryosleep'] = train_df.cryosleep.astype('float64')
test_df['cryosleep'] = test_df.cryosleep.astype('float64')
```

```python
train_df.info()
```

```python
train_df.head()
```

```python
# Total Bill

train_df_total_bill_list = []
for i in range(len(train_df)):
  bill = np.nansum([train_df.iloc[i].roomservice, train_df.iloc[i].foodcourt, train_df.iloc[i].shoppingmall, train_df.iloc[i].spa, train_df.iloc[i].vrdeck])
  train_df_total_bill_list.append(bill)

test_df_total_bill_list = []
for i in range(len(test_df)):
  bill = np.nansum([test_df.iloc[i].roomservice, test_df.iloc[i].foodcourt, test_df.iloc[i].shoppingmall, test_df.iloc[i].spa, test_df.iloc[i].vrdeck])
  test_df_total_bill_list.append(bill)

train_df['total_bill'] = train_df_total_bill_list
test_df['total_bill'] = test_df_total_bill_list
```

```python
services = ['roomservice', 'foodcourt', 'shoppingmall', 'spa', 'vrdeck']

train_df = train_df.drop(columns = services)
test_df = test_df.drop(columns = services)
```

```python
train_df.head()
```

```python
def get_deck(cabin):
  if cabin == 'unknown':
    return np.nan
  else:
    return cabin[0]

train_data_cabin = train_df.cabin.fillna('unknown')
test_data_cabin = test_df.cabin.fillna('unknown')

print(pd.Series(list(map(get_deck, train_data_cabin))).value_counts())
print(pd.Series(list(map(get_deck, test_data_cabin))).value_counts())
```

```python
def deck_to_num(deck):
  if deck == 'unknown':
    return np.nan

  elif deck == 'A':
    return 0

  elif deck == 'B':
    return 1

  elif deck == 'C':
    return 2

  elif deck == 'D':
    return 3

  elif deck == 'E':
    return 4

  elif deck == 'F':
    return 5

  elif deck == 'G':
    return 6

  elif deck == 'T':
    return 7


train_data_decks = list(pd.Series(list(map(get_deck, train_data_cabin))).fillna('unknown'))
test_data_decks = list(pd.Series(list(map(get_deck, test_data_cabin))).fillna('unknown'))

train_df['cabin_deck'] = list(map(deck_to_num, train_data_decks))
test_df['cabin_deck'] = list(map(deck_to_num, test_data_decks))
```

```python
train_df.head()
```

```python
def get_num(cabin):
  if cabin == 'unknown':
    return np.nan

  cabin_num = cabin.split('/')[1]
  return cabin_num

train_data_num_list = list(map(get_num, train_df.cabin.fillna('unknown')))
test_data_num_list = list(map(get_num, test_df.cabin.fillna('unknown')))

train_df['cabin_num'] = train_data_num_list
test_df['cabin_num'] = test_data_num_list
```

```python
train_df.head()
```

```python
def get_side(cabin):
  if cabin == 'unknown':
    return np.nan

  elif cabin[-1] == 'P':
    return 0

  elif cabin[-1] == 'S':
    return 1

train_data_side = list(map(get_side, train_df.cabin.fillna('unknown')))
test_data_side = list(map(get_side, test_df.cabin.fillna('unknown')))

train_df['cabin_side'] = train_data_side
test_df['cabin_side'] = test_data_side
```

```python
train_df.head()
```

```python
train_df = train_df.drop(columns=['cabin'])
test_df = test_df.drop(columns=['cabin'])
```

```python
X = train_df.drop('transported', axis = 1)
y = train_df.transported
```
