```python
def sex_to_float(x):
  if x == 'female':
    return 0.
  else:
    return 1.


data_train = train_df.copy()
# del data_train['Survived']
sex_list = list(data_train['Sex'])
gender_to_float = []

for gender in sex_list:
  if gender == 'female':
    gender_to_float.append(0)
  elif gender == 'male':
    gender_to_float.append(1)
  else:
    gender_to_float.append(np.nan)

gender_to_float

data_train['Sex'] = gender_to_float
data_train.head()
```

```python
data_train.info()
```

```python
embarked_port_to_int = []


for x in data_train.Embarked:
  if x == 'C':
    embarked_port_to_int.append(int(0))

  elif x == 'Q':
    embarked_port_to_int.append(int(1))

  elif x == 'S':
    embarked_port_to_int.append(int(2))

  else:
    embarked_port_to_int.append(np.nan)

data_train['Embarked'] = embarked_port_to_int

data_train.head()


```

```python
data_train.info()
```

```python
data_train = data_train.dropna()
data_train.info()
```

```python
train_target = list(data_train.Survived)
len(train_target)
```

```python
from sklearn.linear_model import LogisticRegression

model = LogisticRegression()
model.fit(data_train, train_target)
```
